```python
# Importa√ß√µes necess√°rias
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# 1. Carregar os dados do Excel
arquivo = '/kaggle/input/dataset/dados.xlsx'
df = pd.read_excel(arquivo)
```

```python
# 2. Agrupar por produto: soma de quantidade vendida e total l√≠quido
ranking_produtos = df.groupby('produto').agg({
    'quantidade': 'sum',
    'valor total liquido': 'sum'
}).reset_index()
```

```python
# 3. Top 10 Produtos por Volume de Vendas
top_vendas = ranking_produtos.sort_values(by='quantidade', ascending=False).head(10)
```

```python
# 4. Top 10 Produtos por Faturamento L√≠quido
top_faturamento = ranking_produtos.sort_values(by='valor total liquido', ascending=False).head(10)
```

```python
# 5. Visualiza√ß√£o com Matplotlib
plt.figure(figsize=(14, 6))
```

```python
# Gr√°fico 1: Mais Vendidos
plt.subplot(1, 2, 1)
plt.barh(top_vendas['produto'], top_vendas['quantidade'], color='dodgerblue')
plt.title('üèÜ Top 10 Produtos Mais Vendidos')
plt.xlabel('Quantidade Vendida')
plt.gca().invert_yaxis()
```

```python
# Gr√°fico 2: Mais Lucrativos
plt.subplot(1, 2, 2)
plt.barh(top_faturamento['produto'], top_faturamento['valor total liquido'], color='seagreen')
plt.title('üí∞ Top 10 Produtos Mais Lucrativos')
plt.xlabel('Faturamento L√≠quido (R$)')
plt.gca().invert_yaxis()

plt.tight_layout()
plt.show()
```

```python
# 6. Mostrar dados em tabelas
print("üèÜ Top 10 Produtos Mais Vendidos (por Quantidade):\n")
print(top_vendas)
print("\nüí∞ Top 10 Produtos Mais Lucrativos (por Faturamento L√≠quido):\n")
print(top_faturamento)
```
